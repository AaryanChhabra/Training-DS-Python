```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
# allows us to make artificial dataset.
```

```python
data = make_blobs(n_samples=200, n_features=2, centers= 4, cluster_std= 1.8, random_state=101)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

```python
# Lets try to find the clusters usi our methods:
```

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=4)
# by blessings, we know here the cluster number
```

```python
kmean.fit(data[0])
```

```python
# Lets compare the real clusters to the kmean predicted one:
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize = (10,6))

ax1.scatter(data[0][:,0], data[0][:,1], c= kmean.labels_)
ax2.scatter(data[0][:,0], data[0][:,1], c= data[1])
```
