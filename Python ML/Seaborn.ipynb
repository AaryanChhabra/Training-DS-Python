```python
# Distribution Plot:

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])

# basically gives a univariant distribution


```

```python

sns.distplot(tips['total_bill'], kde = False, bins = 30) # we removed the fitted line

# kde is kernel density estimation
```

```python
sns.kdeplot(tips["total_bill"])
```

Jointplot:


```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip')

# to combine 2 dist plots, also make a little corelation
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex')
# instead of scatter plot, we can mmake a beautiful hexagon
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg')
# to puy in a  reression line
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde')
```

Plotting Categorical Data

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(data = tips, x = 'sex', y = 'total_bill')

# It shows the mean of the 2 categories.
```

```python
sns.barplot(data = tips, x = 'sex', y = 'total_bill', estimator = np.std)
# like we got mean in the previous figure, we can get any estimator of our choice
```

```python
# Count PLot

sns.countplot(data = tips, x = 'sex')
# Simply counts the number of occurences.
```

```python
# Boxplot

sns.boxplot(data = tips, x = 'day', y = 'total_bill')
```

```python
sns.boxplot(data = tips, x = 'day', y = 'total_bill', hue ='sex')
```

```python
# violin Plot

sns.violinplot(data = tips, x = 'day', y = 'total_bill')

# Similar to box plot, but giving a more information about the distribution.
```

```python

sns.violinplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', split = True)
# After setting hue and split as true we get a very hot looking plot.
```

```python
# Strip Plot:

sns.stripplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', split = True)
```

```python
# Swarm Plot

sns.swarmplot(data = tips, x = 'day', y = 'total_bill')

# Combination of violon plot and the strip plot
# recommended not to use for large dataset
```

```python
sns.swarmplot(data = tips, x = 'day', y = 'total_bill', color = 'black', alpha = 0.5)
sns.violinplot(data = tips, x = 'day', y = 'total_bill')
```

```python
# Factor Plot
# Another way to call the plots we have done till now like mentioning in in the kind.

sns.factorplot(data = tips, x = 'day', y = 'total_bill', kind = 'bar')
```

Matrix Plots:

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
flights.tail()
```

```python
# HeatMap:

# Matrix works as a data for a heatmap:

# Lets make a matrix - simply a correlation between differen variables in the tips dataset.
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot = True, cmap = 'coolwarm')
```

```python
flights.head()
```

```python
# Make a pivot table out of it:

fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, linewidth = 1, linecolor = 'white')
```

```python
sns.clustermap(fp, cmap = 'coolwarm')
# Most similar and months are being clustered together.
```

```python
sns.clustermap(fp, cmap = 'coolwarm', standard_scale= 1)
```

Grids:

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
 sns.pairplot(iris)
```

```python
I = sns.PairGrid(iris)
# Gives more control as compared to pairplot.
I.map_diag(sns.distplot) # makes diagonal as distribution plot
I.map_upper(plt.scatter)
I.map_lower(sns.kdeplot)
```

```python
# Facet grid:
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')

g.map(sns.histplot, 'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')

g.map(plt.scatter, 'total_bill', 'tip')
```

```python
# Regression Plots:

sns.lmplot(x = 'total_bill', y = 'tip', data =tips)
```

```python

sns.lmplot(x = 'total_bill', y = 'tip', data =tips, hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data =tips, row = 'time', col = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data =tips, col = 'day', hue = 'sex')
```
