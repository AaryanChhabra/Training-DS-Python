```python
import matplotlib.pyplot as plt

%matplotlib inline

# allows us  to see the plots we create inside the NB
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)

print(x)

y = x **2

print(y)
```

```python
# Functional method to make a plot:
plt.plot(x,y)

plt.xlabel("LABEL OF X")

plt.title("TITLE")

#plt.show()
```

```python
# putting multiple plots in a same cell:

plt.subplot(2,1,1) # 2 row 1 columns and sub plot number 1
plt.plot(x,y)
plt.ylabel("Y")

plt.subplot(2,1,2) # subplot number 2
plt.plot(y,x)
plt.xlabel("Y")
```

```python
# Object oriented Method:
```

```python
# we create a figure object:

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # These numbers will make sense a little later.

axes.plot(x,y)

axes.set_xlabel("X")

axes.set_title("Title")
```

```python
# Exploring what those number means:

fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1]) # its like fixing the dimension of the figure ([coordinate of the left of the X axis, Coordinate of the bottom of the Y axis, width or the length of the X axis, Height or the length of the Y axis ])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

Part 2

```python
# Creating Sub plots using Object Oriented Programming:

fig,axes = plt.subplots(2,3)

# Automatically makes plots accoring to our input of row and column number.

plt.tight_layout() # removes the problem of overlap among the plots.
```

```python
fig,axes = plt.subplots(2,1)

# As we mentioned axes in the code above, it is axes object is a list of matplotlib axes. We can run a for loop on it.

print(len(axes)) # gives out 2 as ther are 2 axes we defined here.

axes[0].plot(x,y)
axes[1].plot(y,x)

# Of course we can iterate through it like this:

plt.tight_layout()
```

```python
# Setting up the figure size:

# inside of the subplot() and the figure(), we can mention figsize= (x,y)

fig, axes = plt.subplots(1,2, figsize = (8,2))
```

```python
# Managing the pixels and saving the figure:

fig.savefig("Name of the figure.png", dpi = 200)  # increse the dpi to increase the picture quality
```

# Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2)
ax.plot(x,x**3)

# To add a legend:
# We needed to add a label to use in the legend:


ax.plot(x,x**2, label = "X sq")
ax.plot(x,x**3, label = " X cube")

ax.legend(loc = 0)
# We have location code for different positon of te legend, there is a whole documentation available, 0 means the best location.
```

```python
# Adding some beauty to the plots:
# Setting up colors:

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, color = 'orange', linewidth = 2, alpha = 0.5, linestyle = "--", marker = "o", markersize =  10, markeredgecolor = "red", markeredgewidth = "2") # can google RGB HEX code to add fancy colors in the plot.

# marker marks out the points in the graph. we can use [o, +, *] -- There is a huge variety available

# We can just use ls or lw instead of linestyle and linewidth.
```
